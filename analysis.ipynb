# 📓 Exploratory Data Analysis
**Dataset:** SN_d_tot_V2_0.csv — SILSO Daily Sunspot Numbers
**SAP ID:** 70177688

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(
    '../data/SN_d_tot_V2_0.csv', sep=';', header=None,
    names=['Year','Month','Day','DecimalDate','SunspotNumber','StdDev','Observations','Definitive']
)
print('Shape:', df.shape)
df.head(10)
```

```python
# Basic info
df.info()
```

```python
# Replace missing value sentinels
df['SunspotNumber'] = df['SunspotNumber'].replace(-1, np.nan)
df['StdDev']        = df['StdDev'].replace(-1.0, np.nan)
print('Missing values:\n', df.isnull().sum())
```

```python
# Statistical summary
df.describe()
```

```python
# Annual average sunspot number
annual = df.dropna(subset=['SunspotNumber']).groupby('Year')['SunspotNumber'].mean()
plt.figure(figsize=(12,4))
annual.plot(color='#E25822')
plt.title('Annual Average Sunspot Number (1818-2026)')
plt.xlabel('Year'); plt.ylabel('Avg Sunspot Number')
plt.tight_layout(); plt.show()
```

```python
# Distribution
plt.figure(figsize=(8,4))
df['SunspotNumber'].dropna().hist(bins=50, color='#E25822', edgecolor='white')
plt.title('Histogram of Daily Sunspot Numbers')
plt.xlabel('Sunspot Number'); plt.ylabel('Frequency')
plt.tight_layout(); plt.show()
```

```python
# Correlation matrix
num_cols = ['SunspotNumber','StdDev','Observations','Year','Month','Day']
corr = df[num_cols].dropna().corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn')
plt.title('Feature Correlation Matrix')
plt.tight_layout(); plt.show()
```
